```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "C:\\Users\\anubh\\Downloads\\mission_launches.csv"
df = pd.read_csv(file_path)

# Count the number of launches per organization
launch_counts = df["Organisation"].value_counts()

# Create the bar chart
plt.figure(figsize=(12, 6))
launch_counts.plot(kind="bar", color="skyblue", edgecolor="black")

# Labels and title
plt.xlabel("Organization", fontsize=12)
plt.ylabel("Number of Launches", fontsize=12)
plt.title("Number of Space Mission Launches by Organization", fontsize=14)
plt.xticks(rotation=90, ha="right")

# Show the chart
plt.show()
```

```python
import pandas as pd

# Load the CSV file
file_path = "C:\\Users\\anubh\\Downloads\\mission_launches.csv"  # Update with the correct file path
df = pd.read_csv(file_path)

# Convert the Price column to numeric values
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# Calculate the average amount spent per launch by each organization
avg_spent_per_launch = df.groupby("Organisation")["Price"].mean().sort_values(ascending=False)

# Display the results
print(avg_spent_per_launch)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path ="C:\\Users\\anubh\\Downloads\\mission_launches.csv"   # Update with your correct file path
df = pd.read_csv(file_path)

# Extract the year from the Date column
df["Year"] = pd.to_datetime(df["Date"], errors="coerce").dt.year

# Count the number of launches per year
launches_per_year = df["Year"].value_counts().sort_index()

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(launches_per_year.index, launches_per_year.values, marker="o", linestyle="-", color="skyblue")

# Labels and title
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Launches", fontsize=12)
plt.title("Number of Space Mission Launches per Year", fontsize=14)
plt.grid(True)

# Show the chart
plt.show()
```
